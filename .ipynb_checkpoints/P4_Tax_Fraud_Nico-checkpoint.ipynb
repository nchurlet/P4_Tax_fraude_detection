Compte tenu de cette hipothèse, je vais ajouter une colonne `GROUP` qui indiquera :
- 0 pour les non-fraudeurs (`target==0`) 
- 1 pour les fraudeurs (`target!=0`)

Ce sera plus simple pour l'analyse.
Peut-être que je la supprimerai avant la partie modeling.

```python
df_train_initial['GROUP'] = df_train_initial.target.apply(lambda x: 
                               1 if x!=0 
                               else 0)
```

### Réalisation d'un tableau de nan présentant :
- Le code de chaque colonne
- sa description depuis le tableau de description
- le taux de nan par colonnes
- le nombre de valeurs renseignées
- le nombre de valeurs manquantes

- [Création de groupes de features discriminés en fonction du taux de nan et regex](#recueil)

```python
sum_na = df_train_initial.isna().sum()
lines = df_train_initial.shape[0]
df_nans= pd.DataFrame(sum_na, columns=["Number of NaN"])
df_nans["Percents Nan"] = pd.DataFrame(sum_na/lines)
df_nans["Values count"] = lines-sum_na
df_nans = df_nans.merge(
    df_VariableDescription_initial,
    right_index=False,
    left_index=True,
    left_on=df_nans.index,
    right_on=df_VariableDescription_initial['Libellé ']
)
df_nans.drop(['Libellé '], axis=1,inplace=True)
df_nans.sort_values(by="Number of NaN", ascending=False).style.bar(color='#5858FA').set_properties(**{'text-align': 'left'})
```

```python
df_nans.shape
```

La jointure semble avoir bien fonctionné, 121 lignes dans df_nans et 121 colonnes dans le df_initial, toutes les colonnes semblent avoir bien récupéré leur intitulé, cependant: 
- 12 features n'ont pas de descriptions.
- la feature FACAGR n'existe pas dans le dataset initial bien qu'elle soit renseignée dans la description

La target est répartie de la même manière selon les catégories fiscales.
- Notons quand même :
- - la catégorie N est trop faible pour entrer en ligne de compte.
- - Hormis la catégorie N, il y a un léger décalage de la répartition des fraudeurs :
- - - Répartition des fraudeurs de la gauche vers la droite = P < C < M
- Il y a une différence de répartition Fraudeurs / Non- Fraudeurs
- - P : 90 / 10
- - M : 65 / 35
- - C : 80 / 20
- - N : 10 / 90

**Catégorie**|**%Fraudeurs**|**%Non-fraudeurs**
--|--|--
P|90|10
M|65|35
C|80|20
N|10|90

```python
uniques =  df_train_initial['CTR_CATEGO_X'].unique()
for unique, iteration in zip(uniques, range(len(uniques))):
    sns.distplot(
        df_train_initial[df_train_initial['CTR_CATEGO_X']==unique]['target'], 
        color=_colors[iteration%len(uniques)],
         kde_kws = {'bw' : 1.5, 'kernel' : 'tri'}
    )
plt.legend(uniques)
plt.title("Distributions des targets par catégorie de contribuable\n", fontdict=_fontdict)
plt.show()
del uniques
```

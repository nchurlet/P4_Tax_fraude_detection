Compte tenu de cette hipothse, je vais ajouter une colonne `GROUP` qui indiquera :
- 0 pour les non-fraudeurs (`target==0`)

Ce sera plus simple pour l'analyse.

```python
df_train_initial['GROUP'] = df_train_initial.target.apply(lambda x: 
                               1 if x!=0 
                               else 0)
```

### Ralisation d'un tableau de nan prsentant :
- Le code de chaque colonne
- sa description depuis le tableau de description
- le taux de nan par colonnes
- le nombre de valeurs renseignes

- [Cration de groupes de features discrimins en fonction du taux de nan et regex](#recueil)

```python
sum_na = df_train_initial.isna().sum()
lines = df_train_initial.shape[0]
df_nans= pd.DataFrame(sum_na, columns=["Number of NaN"])
df_nans["Percents Nan"] = pd.DataFrame(sum_na/lines)
df_nans["Values count"] = lines-sum_na
df_nans = df_nans.merge(
    df_VariableDescription_initial,
    right_index=False,
    left_index=True,
    left_on=df_nans.index,
    right_on=df_VariableDescription_initial['Libell ']
)
df_nans.drop(['Libell '], axis=1,inplace=True)
df_nans.sort_values(by="Number of NaN", ascending=False).style.bar(color='#5858FA').set_properties(**{'text-align': 'left'})
```
